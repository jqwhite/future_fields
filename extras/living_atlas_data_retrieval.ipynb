## authenticate and generate token

```python
client_id = os.getenv('ARCGIS_CLIENT_ID')
client_secret = os.getenv('ARCGIS_CLIENT_SECRET')

if not client_id or not client_secret:
    raise ValueError("Please set the ARCGIS_CLIENT_ID and ARCGIS_CLIENT_SECRET environment variables")
```

```python
# URL to generate token
token_url = 'https://www.arcgis.com/sharing/rest/oauth2/token/'

# Parameters for token generation
params = {
    'client_id': client_id,
    'client_secret': client_secret,
    'grant_type': 'client_credentials',
    'expiration': 60  # Token expiration in minutes
}

# Request token
response = requests.post(token_url, data=params)
token_info = response.json()
token = token_info['access_token']

print(f'Token: {token}')
```

```python
# Define the URL for the slices endpoint
slices_url = 'https://tiledimageservices3.arcgis.com/0Fs3HcaFfvzXvm7w/arcgis/rest/services/U_S__Climate_Thresholds___LOCA_Historic/ImageServer/slices'

# Define the query payload
payload = {
    "sliceId": 0,
    "multidimensionalDefinition": [
        {
            "variableName": "Consecutive Dry Days",
            "dimensionName": "StdTime",
            "values": [631152060000.0]
        }
    ],
    "f": "json"  # Specify the format as JSON
}

# Make the request
response = requests.post(slices_url, json=payload)

# Check the response
if response.status_code == 200:
    try:
        slice_data = response.json()
        print(slice_data)
    except requests.exceptions.JSONDecodeError:
        print("Error: Received an empty response or non-JSON response")
        print("Response content:", response.text)
else:
    print(f"Error: {response.status_code}")
    print("Response content:", response.text)
```

```python
import requests

# Define the WCS GetCoverage request URL
coverage_id = "LOCA_USA_CONUS_Climate_Thresholds_Consecutive Dry Days 1976-2005 Historical record"
bbox = "-109.0452,31.3322,-103.002,37.0002"  # Bounding box for New Mexico
url = f"https://landscape12.arcgis.com/arcgis/services/LOCA_USA_CONUS_Climate_Thresholds/ImageServer/WCSServer?service=WCS&version=2.0.1&request=GetCoverage&coverageId={coverage_id}&format=image/tiff&subset=Lat({bbox.split(',')[1]},{bbox.split(',')[3]})&subset=Lon({bbox.split(',')[0]},{bbox.split(',')[2]})"

# Send the GetCoverage request
response = requests.get(url)

# Check the response
if response.status_code == 200:
    # Save the coverage data to a file
    with open("coverage_new_mexico.tif", "wb") as file:
        file.write(response.content)
    print("Coverage data saved to 'coverage_new_mexico.tif'")
else:
    print("Error: ", response.text)
```
